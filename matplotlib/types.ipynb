```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python


url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)
print(df.info())
# 1. Line plot — plot average fare per class
#    x = class (1,2,3), y = mean fare
#    Add title, xlabel, ylabel, grid, marker
plt.figure(figsize=(8,4))
line1=df.groupby('Pclass')['Fare'].mean() *100
plt.plot(line1.index,line1.values,marker='o')
plt.title('AVERAGE FARE PER PCLASS')
plt.xlabel('CLASS')
plt.ylabel('AVG FARE')
plt.legend()
plt.grid(True)
plt.show()
# 2. Bar plot — number of passengers per embarkation port
#    S, C, Q — use different color for each bar
#    Rename xticks to 'Southampton', 'Cherbourg', 'Queenstown'
bar1=df.groupby('Embarked')['PassengerId'].count()
plt.figure(figsize=(8,4))
plt.bar(bar1.index,bar1.values,color=['red','blue','green'])
plt.title('NO OF PASSENGERS PER PORT')
plt.xlabel('PORTS')
plt.ylabel('NO OF PASSENGERS')
plt.xticks(['S','C','Q'],['Southampton','Cherbourg','Queenstown'])
plt.show()
# 3. Histogram — distribution of Fare
#    bins=40, color='green'
#    Add a vertical line at mean fare:
#    plt.axvline(df['Fare'].mean(), color='red', linestyle='--', label='Mean')

# 4. Scatter — Age vs Fare
#    Color by Pclass (c=df['Pclass'])
#    cmap='viridis', alpha=0.6

# 5. Create a 2x2 subplot with:
#    [0,0] Age histogram
#    [0,1] Survivors by gender bar plot
#    [1,0] Age vs Fare scatter
#    [1,1] Average fare per class line plot
#    tight_layout, proper titles on all
```
