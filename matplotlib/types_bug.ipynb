```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python



url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

# Bug 1
plt.plot([1,2,3], [10,20,30])
plt.title('Sales')
plt.show()

# Bug 2
plt.figure(figsize=(8,4))
plt.hist(df['Age'].dropna(), bins=30)
plt.show()

# Bug 3
survival = df.groupby('Pclass')['Survived'].sum()
plt.bar(survival.index, survival.values)
plt.xticks([1,2,3],['FIRST','SECOND','THIRD'])
plt.show()

# Bug 4
fig, axes = plt.subplots(1, 2, figsize=(10,4))
axes[0].hist(df['Age'].dropna(), bins=30)
axes[0].set_title('Age Distribution')

axes[1].scatter(df['Age'], df['Fare'], alpha=0.5)
axes[1].set_title('Age vs Fare')
plt.show()


# Bug 5
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].hist(df['Age'].dropna())
axes[1].scatter(df['Age'], df['Fare'])
plt.tight_layout()
plt.show()
```
