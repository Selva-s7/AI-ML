```python
import torch
import torch.nn as nn

class BuggyNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer1 = nn.Linear(3, 6)
        self.layer2 = nn.Linear(6, 1)   # bug 1
        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.relu(self.layer1(x))
        x = self.layer2(x)
        return x

model = BuggyNet()
x = torch.rand(4, 3)   # bug 2
output = model(x)
print("Output shape:", output.shape)
```
