```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('train.csv')
data=data.dropna(subset=['Age','Fare','Pclass','Parch','SibSp'])
X=data[['Age','Fare','Pclass','Parch','SibSp']]
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
pca=PCA(n_components=0.95)
X_pca=pca.fit_transform(X_scaled)
print(pca.n_components_)
print('EXPLAINED VARIENCE RATIO',pca.explained_variance_ratio_)
plt.figure(figsize=(8,4))
plt.scatter(X_pca[:,0],X_pca[:,1],c=data['Survived'],cmap='viridis')
plt.show()
```
