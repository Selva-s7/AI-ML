```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split

messages = [
    "You won a free prize click here",
    "Hello how are you doing today",
    "Free entry win cash prize now",
    "Are you coming to the meeting today",
    "Claim your free reward now",
    "See you at the office tomorrow"
]
labels = [1, 0, 1, 0, 1, 0]

X_train, X_test, y_train, y_test = train_test_split(
    messages, labels, test_size=0.2, random_state=42
)

pipeline = Pipeline(steps=[
    ('tfidf', TfidfVectorizer(stop_words='english')),
    ('model', LogisticRegression())    # bug 1 - duplicate step name
])

pipeline.fit(X_train, y_train)

# bug 2 - manually transforming test data before passing to pipeline

print("Accuracy:", pipeline.score(X_test, y_test))
```
