```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

print("Scikit-learn ready")
```

```python

house_size = [1000,1500,2000,2500,3000,3500,4000,4500,5000]
price =      [200, 300, 400, 500, 600, 700, 800, 900, 1000]
X=np.array(house_size)
Y=np.array(price)
X=X.reshape(-1,1)
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
Mod=LinearRegression()
Mod.fit(X_train,Y_train)
prediction=Mod.predict(X_test)
print('prediction:',prediction)#gives prediction results on new data
print('Actual',Y_test)#actual test data
print('Score',Mod.score(X_test,Y_test))#compares 1.0 useful 0 useless
print(X_test)
print(Y_test)
print("SLOPE",Mod.coef_)
print('INTERCEPT',Mod.intercept_)
```
