```python
import pandas as pd
```

```python
# Create a DataFrame with duplicates
data = {
    'name': ['Alice', 'Bob', 'Alice', 'Diana', 'Bob'],
    'age':  [25, 30, 25, 28, 30]
}
df_dup = pd.DataFrame(data)
#print(df_dup)
# Check duplicates
#print(df_dup.duplicated())  # boolean Series — True where duplicate
#print(df_dup.duplicated().sum())    # how many duplicates
# See the actual duplicate rows
#print(df_dup[df_dup.duplicated()])
# Remove duplicates
df_clean = df_dup.drop_duplicates()
print(df_clean)
print(df_clean.shape)
# Sometimes you want unique by one column only
df_dup.drop_duplicates(subset=['name'])  # keep first occurrence of each name
df_dup.drop_duplicates(subset=['name'], keep='last')  # keep last occurrence



```
