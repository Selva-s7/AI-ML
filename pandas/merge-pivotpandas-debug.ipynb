```python
import pandas as pd

df1 = pd.DataFrame({
    'student_id': [1, 2, 3],
    'name':       ['Alice', 'Bob', 'Charlie']
})

df2 = pd.DataFrame({
    'id':    [1, 2, 4],
    'marks': [95, 80, 70]
})

sales = pd.DataFrame({
    'month':   ['Jan','Jan','Feb','Feb','Mar','Mar'],
    'product': ['A','B','A','B','A','B'],
    'revenue': [100, 200, 150, 250, 120, 300]
})

# Bug 1 — merge on mismatched column names
result = pd.merge(df1, df2, left_on='student_id',right_on='id')
print(result)

# Bug 2 — want all students even without marks
result = pd.merge(df1, df2, left_on='student_id',
                  right_on='id', how='left')
print(result)

# Bug 3 — pivot missing aggfunc
pivot = sales.pivot_table(
    values='revenue',
    index='month',
    columns='product',
    aggfunc='mean'
)
print(pivot)

# Bug 4
pivot = sales.pivot_table(
    values='revenue',
    index='month',
    columns='product',
    aggfunc='sum'
)
print(pivot['A'])      # get column A
print(pivot.loc['Jan',:])    # get row Jan
```
