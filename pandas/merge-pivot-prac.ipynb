```python
import pandas as pd
```

```python
orders = pd.DataFrame({
    'order_id':   [1, 2, 3, 4, 5],
    'customer_id':[101, 102, 101, 103, 102],
    'product':    ['A', 'B', 'A', 'C', 'B'],
    'amount':     [500, 300, 450, 700, 350]
})

customers = pd.DataFrame({
    'customer_id': [101, 102, 104],
    'name':        ['Alice', 'Bob', 'Charlie'],
    'city':        ['Chennai', 'Mumbai', 'Delhi']
})

# 1. Inner merge — which orders have customer info?
result1=pd.merge(orders,customers,on='customer_id',how='inner')
print(result1)
# 2. Left merge — keep all orders, add customer info
#    How many NaN values appear and why?
# 103 has no name and city so 2 Nan
result2=pd.merge(orders,customers,on='customer_id',how='left')
mergeddf=pd.merge(orders,customers,on='customer_id')
# 3. Total amount spent per customer (use merged df)
result3=mergeddf.pivot_table(
    
        values='amount',
        index='name',
        aggfunc='sum'
    
)
#or
#result3=mergeddf.group('name')['amount'].sum()
print(result3.idxmax())
# 4. Which city generated most revenue?
result4=mergeddf.groupby('city')['amount'].sum()
print(result4.idxmax())
# 5. Pivot table — rows=product, values=amount, aggfunc=sum
Pivot=orders.pivot_table(
    values='amount',
    index='product',
    aggfunc='sum'
    
)
print(Pivot)
# 6. Which product made most revenue?
print(Pivot['amount'].idxmax())
```
