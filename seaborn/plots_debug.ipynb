```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python



url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)
df['Age'] = df['Age'].fillna(df['Age'].median())
df = df.drop('Cabin', axis=1)

# Bug 1
corr = df.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap='coolwarm',center=0)
plt.show()

# Bug 2
sns.barplot(data=df,x='Pclass', y='Survived', hue='Sex')
plt.show()

# Bug 3
fig, axes = plt.subplots(1, 2, figsize=(12,4))
sns.histplot(data=df, x='Age', hue='Survived', bins=30,ax=axes[0])
sns.boxplot(data=df, x='Pclass', y='Age',ax=axes[1])
plt.show()

# Bug 4
sns.scatterplot(data=df, x='Age', y='Fare',
                hue='Survived', style='Pclass')
plt.title('AGE VS FARE')
plt.show()

# Bug 5
cols = ['Age', 'Fare', 'Pclass','Survived']
sns.pairplot(df[cols], hue='Survived',diag_kind='hist')
plt.show()
```
