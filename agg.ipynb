```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

print(df.dtypes)
print(df)
# 1. What is the average fare paid per class?
print(df.groupby('Pclass')['Fare'].mean())

# 2. How many passengers embarked from each port?
#    (S, C, Q)
print(df.groupby('Parch')['Embarked'].count())

# 3. What is the survival rate (%) per gender?
#    Expected output:
#    female    74.2%
#    male      18.9%
print(df.groupby('Sex')['Survived'].mean()*100)

# 4. For each class, find:
#    - mean age
#    - total passengers
#    - total survived
#    (use agg with dict)
print(df.groupby('Pclass').agg({'Age':'mean','PassengerId':'sum','Survived':'sum'}))

# 5. Which combination of Pclass + Sex had
#    the highest survival rate?
#    (groupby two columns)
print(df.groupby(['Pclass','Sex'])['Survived'].mean()*100)
```
